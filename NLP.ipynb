## Bag of Words

```python
import pandas as pd
messages=pd.read_csv('SMSSpamCollection',
                    sep='\t',names=["label","message"])
```

```python
messages
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
import re
ps=PorterStemmer()
```

```python
corpus=[]
c=0
for i in range(0,len(messages)):
    review=re.sub('[^a-zA-z]',' ',messages['message'][i])
    review=review.lower()
    review=review.split()
    review=[ps.stem(word) for word in review if not word in stopwords.words('english')]
    review=' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=150)
x = cv.fit_transform(corpus)
x
```

```python
cv.vocabulary_
```

```python
cv = CountVectorizer(max_features=500,ngram_range=(1,2))
x = cv.fit_transform(corpus)
cv.vocabulary_
```

```python
corpus=[]
c=0
for i in range(0,len(messages)):
    review=re.sub('[^a-zA-z]',' ',messages['message'][i])
    review=review.lower()
    review=review.split()
    review=[lemmatizer.lemmatize(word) for word in review if not word in stopwords.words('english')]
    review=' '.join(review)
    corpus.append(review)
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
```

```python
words=[]
for sent in corpus:
    sent_token = sent_tokenize(sent)
    if(not sent_token):
        print(sent_token)
    for sen in sent_token:
        words.append(simple_preprocess(sen))
```

```python
len(words),len(corpus)
```

```python
words
```

```python
import gensim
model=gensim.models.Word2Vec(words)
```

```python
model.wv.index_to_key
```

```python
model.wv.similar_by_word('good')
```

```python
import numpy as np
def AvgWord2Vec(doc):
    return np.mean([model.wv[word] for word in doc if word in model.wv.index_to_key],axis=0)
```

```python
x=[]
for i in range(len(words)):
    x.append(AvgWord2Vec(words[i]))
```
